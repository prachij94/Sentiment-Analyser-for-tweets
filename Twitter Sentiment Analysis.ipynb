We use *AppAuthHandler* rather than mostly used- *OAuthHandler* for authorizing the API. OAuthHandler allows a maximum of 180 requests per 15-minute window(and a maximum of 100 tweets per request). But AppAuthHandler has its maximum request rate limit as 450 and 100 tweets per request. Thus it helps in fetching 45,000 tweets in fifteen minutes. 

```python
auth = tweepy.AppAuthHandler(consumer_key,consumer_secret)
```

```python
api = tweepy.API(auth)
```

Fetching and saving the tweets for a search keyword into a text file

```python
import jsonpickle
```

```python
filename = 'tweetsdownloaded.txt'
searchquery = '#DelhiRiots2020'
max_tweets = 100000000
since_id = None
max_id = -1
tweet_count = 0
```

```python
with open(filename,'w') as f:
    print("Downloading hashtag" + searchquery)
    while(tweet_count<max_tweets):
        try:
            if(max_id<=0):
                if(not since_id):
                    new_tweets = api.search(q=searchquery,count=100,lang="en",tweet_mode='extended')

                else:
                    new_tweets = api.search(q=searchquery,count=100,lang="en",tweet_mode='extended',since_id=since_id)
            else:
                if(not since_id):
                    new_tweets = api.search(q=searchquery,count=100,lang="en",tweet_mode='extended',max_id=str(max_id-1))
                else:
                    new_tweets = api.search(q=searchquery,count=100,lang="en",tweet_mode='extended',max_id=str(max_id-1),since_id=since_id)

            # Tweets Exhausted
            if(not new_tweets):
                print("No more tweets found!!")
                break
            # write all the new_tweets to a json file
            for tweet in new_tweets:
                f.write(jsonpickle.encode(tweet._json,unpicklable=False)+'\n')
                tweet_count+=len(new_tweets)
                print("Successfully downloaded {0} tweets".format(tweet_count))
                max_id=new_tweets[-1].id
        # in case of any error
        except tweepy.TweepError as e:
                print("Some error!!:"+str(e))
                break
```

From the tweets downloaded in the .txt file above, keeping only the required metadata from the json objects i.e. *full_text(full text of the tweet), retweet_count(of a tweet), favorite_count(of a tweet), geo(geographic location), coordinates(latitude and longitude coordinates of the place ), place(from which it is tweeted), follower_count(of the user who tweeted), created_at(time and date of tweet), id_str(id of the tweet in a string format)* and discarding the rest

```python
import json
import csv


f = open('downloadedtweets.csv','a',encoding='utf-8')
csvWriter = csv.writer(f)
csvWriter.writerow(['full_text','retweet_count','user_followers_count','favorite_count','place','coordinates','geo','created_at','id_str'])
```

Now reading the downloaded csv file to clean it up for unwanted keywords like stopwords,etc. Read the csv with encoding parameter as ‘unicode_escape’ in read_csv() function as many tweets contain emojis which can distort the analysis

```python
import pandas as pd
df = pd.read_csv('downloadedtweets.csv', encoding = 'unicode_escape')
df.head()
```

Removing duplicate tweets and dropping columns with mostly null values.

```python
df=df.drop_duplicates(['full_text'])
df.reset_index(drop=True,inplace=True)
df.head()
```

```python
df=df.drop(['place','coordinates','geo','id_str'],axis=1)
df.head()
```

```python
print(df['full_text'][:15])
```

```python
import re
for i in range(len(df)):
    txt = df.loc[i]["full_text"]
    txt=re.sub(r'@[A-Z0-9a-z_:]+','',txt)#replace username-tags
    txt=re.sub(r'^[RT]+','',txt)#replace RT-tags
    txt = re.sub('https?://[A-Za-z0-9./]+','',txt)#replace URLs
    txt=re.sub("[^a-zA-Z]", " ",txt)#replace hashtags
    df.at[i,"full_text"]=txt
```

```python
df.head()
```

```python
import nltk
from nltk.corpus import stopwords
```

```python
stop_words = stopwords.words('english')
for i in range(len(df.index)):
    text = df.loc[i]['full_text']
    tokens = nltk.word_tokenize(text)
    tokens = [word for word in tokens if word not in stop_words]
    
    tokens_sent=' '.join(tokens)
    df.at[i,"tokenised_tweet"] = tokens_sent
```

```python
df.head()
```

Sentiment Analysis is understanding and extracting feelings from data

-1 < sentiment < 1

Polarity measures how positive or negative some text is

subjectivity measures how much of it is an opinion versus factual

```python
for f in range(len(df.index)):
    text = df.loc[f]['tokenised_tweet']
    
    df.at[f,"tweet_polarity"] = TextBlob(text).sentiment.polarity
    df.at[f,"tweet_subjectivity"] = TextBlob(text).sentiment.subjectivity
    if TextBlob(text).sentiment.polarity > 0:
        df.at[f,"Sentiment"] = 'Positive'
    else:
        df.at[f,"Sentiment"] = 'Negative'
```
